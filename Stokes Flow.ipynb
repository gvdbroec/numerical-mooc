L'objectif est de résoudre l'équation de Navier-Stokes pour un écoulement incompressible dans une cavité carrée :
$$\frac{\partial{u}}{\partial{t}} + u\ . \nabla{u} = -\frac{1}{\rho}\ . \nabla{p} + \nu\ . \nabla^2{u}$$

où $u$ est la vitesse, $t$ le temps, $\rho$ la masse volumique du fluide, $p$ la pression et $\nu$ la viscosité du fluide.

Nous allons d'abord simplifier le problème en linéarisant l'équation.

$\color{blue}{\underline{\large{\textbf{Vorticity}}}}$

Le modèle Stokes-flow est efficace pour des écoulements ayant un nombre de Reynolds très petit. En considérant ce nombre proche de 0, l'équation devient :

$$0 = -\nabla{p} + \nabla^2{u}$$

On applique le rotationnel à chaque côté de l'équation :

$$\nabla \times 0 = (\nabla \times -\nabla{p}) + (\nabla \times \nabla^2{u})\ \ \ \Longleftrightarrow \ \ \ \nabla \times \nabla^2{u} = 0$$

On définit $\omega$ la $vorticité$ par : 
$$\omega = \nabla \times u\ \ \ \Longleftrightarrow\ \ \ \nabla \times \nabla^2{u} = \nabla^2{\omega}$$

On obtient finalement une équation de Stokes simplifiée :

$$\nabla^2{\omega} = 0 \ \ \ \ \ \ \ \ \ \ (1)$$

$\color{blue}{\underline{\large{\textbf{Stream function}}}}$

Introduisons la fonction de courant $\psi$ tel que :
$$ u = \frac{\partial{\psi}}{\partial{y}}\ \ et\ \ \nu = -\frac{\partial{\psi}}{\partial{x}} $$

En 2 dimensions, la vorticité $\omega$ peut s'écrire comme : 

$$\omega = \frac{\partial{\nu}}{\partial{x}} - \frac{\partial{u}}{\partial{y}} = -\frac{\partial^2{\psi}}{\partial{x^2}} - \frac{\partial^2{\psi}}{\partial{y^2}}\ \ \ \Longleftrightarrow\ \ \ \nabla^2{\psi} = -\omega \ \ \ \ \ \ \ \ \ \ (2)$$

Les équations (1) et (2) couplées décrivent l'écoulement d'un fluide dans une cavité à couvercle pour des nombres de Reynolds très bas :


$$\left\{ \begin{array}{11} \nabla^2{\omega} = 0 \ \  \ \ \ \ \ \ \ \ \ \ \ (1)\\ \nabla^2{\psi} = -\omega \ \ \ \ \ \ \ \ \ \ (2)\end{array}\right.$$


$\\$

Nous supposons que le couvercle bouge à une vitesse $u = 1$ et qu'il n'y a aucune fuite du liquide en dehors de la cavité. Nous voulons visualiser à quoi ressemble le champ d'écoulement à l'intérieur de la cavité à l'état stable.

$\color{blue}{\underline{\large{\textbf{Boundary conditions}}}}$

Les conditions de bords sont décrites en fonction de $\psi$ et de ses dérivées, comme indiqué sur la figure suivante :
<img src="./figures/drivencavity.svg" width=400px>

Pour l'équation de Laplace : $\nabla^2{\omega}=0$, nous devons représenter les conditions de bords pour $\omega$ en fonction de $\psi$.
Il ne peut pas y avoir de vitesse dans la direction $y$ car la surface est solide. Ainsi,$\frac{\partial^2{\psi}}{\partial{x^2}}=0$ et l'équation (2) devient :
$$\nabla^2{\psi} = -\omega \ \ \ \Longleftrightarrow\ \ \ \frac{\partial^2{\psi}}{\partial{x^2}} + \frac{\partial^2{\psi}}{\partial{y^2}} = -\omega \ \ \ \Longleftrightarrow\ \ \ \frac{\partial^2{\psi}}{\partial{y^2}} = -\omega$$

En faisant une discrétisation centrale de $2^e$ ordre, on a :

$$\omega_{\ j} = -\left ( \frac{\psi_{\ j+1}-2\psi_{\ j}+\psi_{\ j-1}}{\Delta{y}^2} \right )$$

La valeur $\psi_{\ j+1}$ étant hors du domaine, nous devons discrétiser $\frac{\partial{\psi}}{\partial{y}}$ au $3^e$ ordre pour obtenir une valeur de $\psi_{\ j+1}$ qui dépend seulement de points faisant partie du domaine.

$$ \left. \frac{\partial{\psi}}{\partial{y}} \right |_{\ j} = \frac{2\psi_{\ j+1}+3\psi_{\ j}-6\psi_{\ j-1}+\psi_{\ j-2}}{6\Delta{y}}$$

Mais $\left. \frac{\partial{\psi}}{\partial{y}} \right |_{\ j} = u_{\ j}$. On obtient alors :
$$\psi_{\ j+1} = \frac{6\Delta{y}\ u_{\ j}-3\psi_{\ j}+6\psi_{\ j-1}-\psi_{\ j-2}}{2}$$

Où $u_{\ j} = \left \{ \begin{array}{1 1} 1 \ \ \ \mbox{pour le bord supérieur} \\ 0 \ \ \ \mbox{pour les 3 autres bords} \end{array} \right .$

On a donc une condition le long de chaque bord pour $\omega$ en fonction de $\psi$. $\\$

Pour le bord supérieur :
$$\omega_{\ i,\ j} = -\frac{1}{2\Delta{y}^2} \left ( 8\psi_{\ i,\ j-1}-\psi_{\ i,\  j-2} \right ) - \frac{3}{\Delta{y}} + \mathcal{O}(\Delta{y}^2) $$

Pour le bord inférieur :
$$\omega_{\ i,\ j} = -\frac{1}{2\Delta{y}^2} \left ( 8\psi_{\ i,\ j+1}-\psi_{\ i,\  j+2} \right ) + \mathcal{O}(\Delta{y}^2) $$

Pour le bord droit :
$$\omega_{\ i,\ j} = -\frac{1}{2\Delta{y}^2} \left ( 8\psi_{\ i-1,\ j}-\psi_{\ i-2,\  j} \right ) + \mathcal{O}(\Delta{y}^2) $$

Pour le bord gauche :
$$\omega_{\ i,\ j} = -\frac{1}{2\Delta{y}^2} \left ( 8\psi_{\ i+1,\ 1}-\psi_{\ i+2,\  j} \right ) + \mathcal{O}(\Delta{y}^2) $$

$\color{blue}{\underline{\large{\textbf{Méthode de Jacobi}}}}$

Nous allons employer la méthode de Jacobi pour résoudre ces équations. 
$\\$

L'équation (1) est une équation de Laplace, nous avons donc :

$$\nabla^2{\omega} = 0 \ \ \ \Longleftrightarrow\ \ \ \frac{\partial^2{\omega}}{\partial{x}^2} + \frac{\partial^2{\omega}}{\partial{y}^2} = 0$$

En faisant une discrétisation centrale de second ordre et comme $\Delta{x} = \Delta{y}$, on obtient :

$$\omega_{\ i+1,\ j} + \omega_{\ i-1,\ j} + \omega_{\ i,\ j+1} + \omega_{\ i,\ j-1} - 4\omega_{\ i,\ j} = 0$$

La méthode de Jacobi permet de remplacer la valeur d'un point de la grille par la moyenne des 4 points autour de celui-ci. Nous avons alors :

$$\omega_{\ i,\ j}^{\ k+1} = \frac{1}{4} \left ( \omega_{\ i,\ j-1}^{\ k} + \omega_{\ i,\ j+1}^{\ k} + \omega_{\ i-1,\ j}^{\ k} + \omega_{\ i+1,\ j}^{\ k} \right )$$

Par contre l'équation (2) est une équation de Poisson, il y a donc un terme de source qu'il faut prendre en compte.
$$\nabla^2{\psi} = -\omega$$

Même raisonnement que précédemment, et on obtient une équation pour $\psi$ au point $i,\ j$ en fonction des points voisins :
$$\psi_{\ i,\ j}^{k+1} = \frac{ \psi_{\ i+1,\ j}^{k} + \psi_{\ i-1,\ j}^{k} + \psi_{\ i,\ j+1}^{k} + \psi_{\ i,\ j-1}^{k} + \omega_{ i, j}^{k} \ \Delta{x}^2}{4}$$

$\\$

$\underline{\textit{Remarque}}$ : Les indices $i$ et $j$ correspondent aux coordonnées des points de la grille. Et l'indice $k$ correspond au nombre d'itérations réalisées.

$\\$

$\color{blue}{\underline{\large{\textbf{Code}}}}$

Importons d'abord les librairies habituelles et introduisons les valeurs initiales.

```python
from matplotlib import pyplot
import numpy
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16

from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

# Valeurs initiales
nx = 41
ny = 41

l = 1.
h = 1.

x = numpy.linspace(0,l,nx)
y = numpy.linspace(0,h,ny)


dx = l/(nx-1)
dy = h/(ny-1)
```

Nous allons résoudre les équations (1) et (2) simultanément jusqu'au moment où la norme L1 de la différence entre les itérations successives est inférieure à $1E^{-6}$ pour les 2 quantités.

Définissons cette norme L1 :

```python
L1target = 1e-6

def L1norm(new, old) :
    norm = numpy.sum(numpy.abs(new-old))
    return norm
```

Créons maintenant la fonction utilisant la méthode de Jacobi et prenant en compte les conditions initiales et les conditions de bords de notre problème. Quand la condition pour les normes L1 de $\omega$ et de $\psi$ sera atteinte, cette fonction nous donnera :

1) les valeurs de $\omega$ et de $\psi$ sur tout le domaine à la dernière itération

2) les normes L1 de $\omega$ et de $\psi$

3) le nombre d'itérations effectuées.

```python
w = numpy.zeros((nx,ny))
p = numpy.zeros((nx,ny))

def function(w, p, L1target, dy) :

    L1w = 1.
    L1p = 1.
    
    k = 0  # k est le nombre d'itérations   
    
    while not (L1w < L1target and L1p < L1target) :
        
        w_k = w.copy()
        p_k = p.copy()

        
        # Valeur de omega et psi avec la méthode de Jacobi
        w[1:-1,1:-1] = 0.25 * (w_k[1:-1,:-2] + w_k[1:-1,2:] + w_k[:-2,1:-1] + w_k[2:,1:-1])
        p[1:-1,1:-1] = 0.25 * (p_k[2:,1:-1] + p_k[:-2,1:-1] + p_k[1:-1,2:] + p_k[1:-1,:-2] + (w_k[1:-1,1:-1]*(dx**2)))

                
        # Conditions de bords pour omega et psi
        w[-1,:] = (-0.5/(dy**2))*((8.*p[-2,:])-(p[-3,:]))-(3./dy) # Bord supérieur
        p[-1,:] = 0.
        
        w[0,:] = (-0.5/(dy**2))*((8.*p[1,:])-(p[2,:]))        # Bord inférieur
        p[0,:] = 0.
                
        w[:,-1] = (-0.5/(dy**2))*((8.*p[:,-2])-(p[:,-3]))        # Bord droit
        p[:,-1] = 0.
        
        w[:,0] = (-0.5/(dy**2))*((8.*p[:,1])-(p[:,2]))        # Bord gauche
        p[:,0] = 0.
                
              
        # Norme pour omega et psi
        L1w = L1norm(w, w_k)
        L1p = L1norm(p, p_k)
        
        
        # Nombre d'itérations qui seront effectuées
        k = k+1
        
    return w, p, L1w, L1p, k

w, p, L1w, L1p, k = function(w, p, L1target, dy)
```

$\color{blue}{\underline{\large{\textbf{Résultats}}}}$

Vérifions que les normes L1 de $\omega$ et de $\psi$ sont effectivement inférieures à $1E^{-6}$.

```python
print ("La norme L1 de la vorticité, ω, est :", L1w)
print ("La norme L1 de la fonction de courant, ψ, est :", L1p)
```

Regardons combien d'itérations ont été effectuées pour résoudre le problème avec la méthode de Jacobi

```python
print ("Le nombre d'itérations effectuées est :", k)
```

Regardons maintenant à quoi correspond notre situation finale dans un graphique à 2 dimensions puis dans un graphique à 3 dimensions.

```python
# Définissons d'abord la fonction plot_3D qui va nous permettre de représenter notre situation finale en 3 dimensions

def plot_3D(x, y, z) :

    fig = pyplot.figure(figsize=(11,7), dpi=100)
    ax = fig.gca(projection='3d')
    X,Y = numpy.meshgrid(x,y)
    surf = ax.plot_surface(X,Y,z[:], rstride=1, cstride=1, cmap=cm.viridis,
            linewidth=0, antialiased=False)

    ax.set_xlim(0,1)
    ax.set_ylim(0,1)
    ax.set_xlabel('$x$', fontsize = 24)
    ax.set_ylabel('$y$', fontsize = 24)
    ax.set_zlabel('$z$', fontsize = 24)
    ax.view_init(30,45)
    ax.tick_params(axis='both',labelsize=12)
```

```python
pyplot.figure(figsize=(7,7))
pyplot.contourf(x,y,w,10,cmap=cm.viridis)
pyplot.title("Contour plot of vorticity $\omega$ \n at steady state", fontsize = 20)
pyplot.xlabel('$x$', fontsize=24)
pyplot.ylabel('$y$', fontsize=24)
pyplot.tick_params(axis='both',labelsize=12)

plot_3D(x, y, w)
```

```python
pyplot.figure(figsize=(7,7))
pyplot.contourf(x,y,p,10,cmap=cm.viridis)
pyplot.title("Contour plot of streamfunction $\psi$ \n at steady state", fontsize = 20)
pyplot.xlabel('$x$', fontsize=24)
pyplot.ylabel('$y$', fontsize=24)
pyplot.tick_params(axis='both',labelsize=15)

plot_3D(x, y, p)
```
